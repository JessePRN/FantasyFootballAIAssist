# NFL Data Munging and Transformation 

```python
#import pandas in to our code so we can use it as well as plotly and merging 2 csv on a left join. 
import pandas as pd 
import plotly.express as px  
import missingno as msno
data1 = pd.read_csv('data/NFL_3.csv') 
data2 = pd.read_csv('data/NFL_2.csv') 
data3 = pd.read_csv('data/NFL_1.csv') 
df = pd.merge(data1, data2, 
                   on = 'Player',
                   how ='left')
df.head(5)
```

```python
# Using merge on a left join to merge the remaining csv with the newly merge df.
nfl_df = pd.merge(df, data3, 
                   on = 'Player',
                   how ='left')
nfl_df.head(10)
```

```python
# Displaying columns
nfl_df.columns
```

```python
# Dropping redundant and obselete columns and targeting features that will be useful in machine learning. 
nflc_df = nfl_df.drop(['Unnamed: 0_x','Int','Fumbles','FumblesLost','Year ','Unnamed: 0_y','Tm_x','Age_x','Int_x','Fumbles_x','Pos', 'Pos_y', 'FumblesLost_x','Year_x','Tm_y','Age_y','Int_y','Fumbles_y','FumblesLost_y','Year_y'], axis=1)
nflc_df.head(10)
```

```python
# Parsing out arcane columns to more meaningly names by year so that data is in columns by player and not in rows. 
nflcleaned_df = nflc_df.rename(columns ={'Pos_x':'Pos','G':'2021 Games','GS':'2021 GS','Tgt': '2021 Tgt', 'Rec': '2021 Rec', 'PassingYds': '2021 PassingYds' , 'PassingTD': '2021 PassingTD', 'PassingAtt': '2021 PassingAtt', 'RushingYds': '2021 RushingYds', 'RushingTD': '2021 RushingTD', 'RushingAtt': '2021 RushingAtt', 'ReceivingYds': '2021 ReceivingYds', 'ReceivingTD': '2021 ReceivingTD', 'FantasyPoints': '2021 FantasyPoints', 'G_x':'2019 Games','GS_x':'2019 GS', 'Tgt_x':'2019 Tgt', 'Rec_x':'2019 Rec', 'PassingYds_x': '2020 PassingYds', 'PassingTD_x': '2019 PassingTD', 'PassingAtt_x': '2019 PassingAtt', 'RushingYds_x': '2019 RushingYds', 'RushingTD_x': '2019 RushingTD', 'RushingAtt_x': '2019 RushingAtt', 'ReceivingYds_x': '2019 ReceivingYds', 'ReceivingTD_x': '2019 ReceivingTD','FantasyPoints_x': '2019 FantasyPoints', 'G_y': '2020 Games', 'GS_y':'2020 GS', 'Tgt_y': '2020 Tgt', 'Rec_y': '2020 Rec', 'PassingYds_y': '2020 PassingYds', 'PassingTD_y': '2020 PassingTD', 'PassingAtt_y': '2020 PassingAtt', 'RushingYds_y': '2020 RushingYds', 'RushingTD_y': '2020 RushingTD', 'RushingAtt_y': '2020 RushingAtt', 'ReceivingYds_y': '2020 ReceivingYds', 'ReceivingTD_y': '2020 ReceivingTD', 'FantasyPoints_y': '2020 FantasyPoints'})
nflcleaned_df.head(10)
```

```python
# Shape is just an attribute that tells us the shape of our DataFrame in the format (rows, columns)
print(nflcleaned_df.shape) 
print('We have', nflcleaned_df.shape[0], 'players we can analyze for the 2019, 2020, 2021 seasons!')
print('We have', nflcleaned_df.shape[1], 'columns of data we can analyze for the 2019, 2020, 2021 seasons!')
```

# NFL Players Running Back Data

```python
# Running Back df with engineered features parsed out by year
rb_df = nflcleaned_df.loc[nflcleaned_df['Pos'] == 'RB', ['Player', 'Tm', 'Age', '2021 Games', '2021 GS', '2021 Tgt',
       '2021 Rec','2021 RushingYds', '2021 RushingTD', '2021 RushingAtt',
       '2021 ReceivingYds', '2021 ReceivingTD', '2021 FantasyPoints',
       '2019 Games', '2019 GS', '2019 Tgt', '2019 Rec','2019 RushingYds',
       '2019 RushingTD', '2019 RushingAtt', '2019 ReceivingYds',
       '2019 ReceivingTD', '2019 FantasyPoints', '2020 Games', '2020 GS',
       '2020 Tgt', '2020 Rec','2020 RushingYds', '2020 RushingTD',
       '2020 RushingAtt', '2020 ReceivingYds', '2020 ReceivingTD',
       '2020 FantasyPoints']]
rb_df['Usage19'] = ((nflcleaned_df['2019 Tgt'] + nflcleaned_df['2019 RushingAtt']) / nflcleaned_df['2019 Games'])
rb_df['Touchdowns19'] = nflcleaned_df['2019 RushingTD'] + nflcleaned_df['2019 ReceivingTD']
rb_df['TotYdsScrm19']= nflcleaned_df['2019 ReceivingYds']+ nflcleaned_df['2019 RushingYds']
rb_df['Production19'] = ((nflcleaned_df['2019 RushingTD'] + nflcleaned_df['2019 ReceivingTD'] + nflcleaned_df['2019 RushingYds'] + nflcleaned_df['2019 ReceivingYds']) / rb_df['Usage19'])
rb_df['Usage20'] = ((nflcleaned_df['2020 Tgt'] + nflcleaned_df['2020 RushingAtt']) / nflcleaned_df['2020 Games'])
rb_df['Touchdowns20'] = nflcleaned_df['2020 RushingTD'] + nflcleaned_df['2020 ReceivingTD']
rb_df['TotYdsScrm20']= nflcleaned_df['2020 ReceivingYds']+ nflcleaned_df['2020 RushingYds']
rb_df['Production20'] = ((nflcleaned_df['2020 RushingTD'] + nflcleaned_df['2020 ReceivingTD'] + nflcleaned_df['2020 RushingYds'] + nflcleaned_df['2020 ReceivingYds']) / rb_df['Usage20'])
rb_df['Usage21'] = ((nflcleaned_df['2021 Tgt'] + nflcleaned_df['2021 RushingAtt']) / nflcleaned_df['2021 Games'])
rb_df['Touchdowns21'] = nflcleaned_df['2021 RushingTD'] + nflcleaned_df['2021 ReceivingTD']
rb_df['TotYdsScrm21']= nflcleaned_df['2021 ReceivingYds']+ nflcleaned_df['2021 RushingYds']
rb_df['Production21'] = ((nflcleaned_df['2021 RushingTD'] + nflcleaned_df['2021 ReceivingTD'] + nflcleaned_df['2021 RushingYds'] + nflcleaned_df['2021 ReceivingYds']) / rb_df['Usage21'])
rb_df.head(10)
```

```python
msno.bar(rb_df)
```

```python
rb_df.fillna(0, inplace=True)
```

```python
# Using previous created stats to create an average of all three years if played. 
rb_df['Average Total Usage'] = ((rb_df['Usage19'] + rb_df['Usage20'] + rb_df['Usage21']) / 3)
rb_df['Average Total Yards'] = ((rb_df['TotYdsScrm19'] + rb_df['TotYdsScrm20'] + rb_df['TotYdsScrm21']) / 3)
rb_df['Average Total Production'] = ((rb_df['Production19'] + rb_df['Production20'] + rb_df['Production21']) / 3)
rb_df.head(10)
```

```python
# Plotting the data
from matplotlib import pyplot as plt

plt.style.use('ggplot')

rb_df.plot(x='Average Total Usage', y='Average Total Production', kind='scatter')
```

```python
rb_df.to_csv('data/RB_X')
```
